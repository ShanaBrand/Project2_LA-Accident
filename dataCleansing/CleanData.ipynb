```python
import pandas as pd
import csv
import numpy as np
import time as time
from datetime import datetime,date, time
```

```python
path = "RAW_trafficData.csv" #LA Traffic collison csv file 
```

```python
data = pd.read_csv(path, encoding="UTF-8")
```

```python
data.head()
```

```python
#COLUMNS RETAINED: Date Occurred, Time Occurred, Victim Age, Victim Sex, Victim Descent, Location
extract = data[["DR Number","Date Occurred", "Time Occurred", "Victim Age", "Victim Sex", "Victim Descent", "Location","Area Name","Address","Cross Street"]]
```

```python
extract1 = extract.rename(columns={"DR Number":"dr_number","Date Occurred":"date_occurred", "Time Occurred":"time_occurred","Victim Age":"victim_age","Victim Sex":"victim_sex","Victim Descent":"victim_descent","Location":"location","Area Name":"area_name","Address":"address","Cross Street":"cross_street"})
```

```python
extract1["cross_street"].value_counts()
```

```python
extract1.dtypes
```

```python
extract1["intersection"] = extract1["address"]+ " / " +extract1["cross_street"]
```

```python
extract1["intersection"].count()
```

```python
extract1["time_occurred"].max()
```

```python
extract1["date_occurred"]=pd.to_datetime(extract1["date_occurred"])
```

```python
#extract1["time_occurred"].astype(str)
```

```python
split_data = extract1.location.str.strip(')').str.strip("(").str.split(', ')
```

```python
extract1['latitude'] = split_data.apply(lambda x: x[0])
extract1['longitude'] = split_data.apply(lambda x: x[1])
```

```python
extract1.head()
```

```python
extract1.time_occurred
```

```python
extract1.head()
```

```python
extract1["year_occur"] = extract1["date_occurred"].dt.year
extract1["month_occur"] = extract1["date_occurred"].dt.month
extract1["day_occur"] = extract1["date_occurred"].dt.day
```

```python
extract1.head()
```

```python
#How Many Null values are there
extract1.isnull().sum()
```

```python
#Check for DUPLICATES and count them 
extract1.duplicated().sum()
```

```python
extract1.shape #How many rows and columns are in the dataset
```

```python
#Drop all rows with null values
drop_na = extract1.dropna(how='any')
```

```python
drop_na.shape
```

```python
drop_na.duplicated().sum()
```

```python
#Drop all rows with duplicate values
drop_na_dup = drop_na.drop_duplicates()
```

```python
drop_na_dup.shape
```

```python
drop_na_dup.head()
```

```python
clean_extract = drop_na_dup[["dr_number","date_occurred","year_occur","month_occur","day_occur","time_occurred", "victim_age", "victim_sex", "victim_descent", "latitude", "longitude","area_name","intersection"]]
```

```python
clean_extract.head()
```

```python
clean_extract.victim_age = clean_extract.victim_age.astype(int)
```

```python
clean_extract.head()
```

```python
clean_extract.to_csv("LAtraffic_clean.csv",header=True,index=False)
```
